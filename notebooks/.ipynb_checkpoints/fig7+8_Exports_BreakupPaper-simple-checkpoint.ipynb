```python
import matplotlib.pylab as plt
import matplotlib.pyplot as pyp
import numpy as npy
import sys

def sub_create(xax,yax,xgap=0.05,ygap=0.05,xstart=-1,ystart=-1,xend=-1,yend=-1,xstep=20,ystep=10,xlabel='',ylabel='',xh=None,yh=None,rm_labels=1,figscale=2,scalex=0,scaley=0,proj=0):

  if len([proj])==1:
    proj_per_plot = proj*npy.ones([xax*yax,1])
  elif len(proj) < xax*yax:
    print('wrong amount of projections provided - exiting')
    sys.exit()
  else:
    proj_per_plot = proj*1

  if xh == None:
    xh = npy.ones(xax)
  if yh == None:
    yh = npy.ones(yax)
  else:
    yh = yh[::-1]

  ##Error handling
  if len(xh) != xax:
    print('please provide an x-widths array of length '+str(xax))
    sys.exit()
  if len(yh) != yax:
    print('please provide an y-heights array of length '+str(yax))
    sys.exit()

  fig1 = plt.figure(figsize=(xax*(figscale+scalex)+1,yax*(figscale+scaley)+1))

  xwide = (1-xgap*(1+xax))/npy.sum(xh)
  ywide = (1-ygap*(1+yax))/npy.sum(yh)

  ax = []
  loop_cnt = 1
  for y in npy.arange(yax-1,-1,-1):
    for x in range(xax):
      xloc = xgap*(x+1) + xwide*npy.sum(xh[0:x])
      yloc = ygap*(y+1) + ywide*npy.sum(yh[0:y])
      addaxes = [xloc,yloc,xwide*xh[x],ywide*yh[y]]
      if proj_per_plot[loop_cnt-1] == 1:
        ax.append(fig1.add_axes(addaxes,projection='polar'))
        print('polar '+str(loop_cnt))
      elif proj_per_plot[loop_cnt-1] == 2:
        ## Polar plot, but with subset, not full Arctic
        import mpl_toolkits.axisartist.angle_helper as angle_helper
        from matplotlib.projections import PolarAxes
        from matplotlib.transforms import Affine2D
        from mpl_toolkits.axisartist import SubplotHost
        from mpl_toolkits.axisartist import GridHelperCurveLinear
        tr = Affine2D().scale(npy.pi/180,1.) + PolarAxes.PolarTransform()
        extreme_finder = angle_helper. ExtremeFinderCycle(20,20,lon_cycle=360,lat_cycle='None',
          lon_minmax=None,lat_minmax=(0,npy.inf))
        grid_locator1 = angle_helper.LocatorDMS(12)
        tick_formatter1 = angle_helper.FormatterDMS()
        grid_helper = GridHelperCurveLinear(tr,extreme_finder=extreme_finder,
                                            grid_locator1=grid_locator1,tick_formatter1=tick_formatter1)

        ax1 = SubplotHost(fig1, xax, yax, loop_cnt, grid_helper=grid_helper)
        fig1.add_axes(ax1) #fig.add_subplot(ax1)
        ax1.axis["lat"] = axis = ax1.new_floating_axis(0,60)
        ##NEED TO IMPLEMENT THIS
        #axis.label.set_text(r"$\theta = 60^{\circ}$")
        #axis.label.set_visible(True)

        ax1.axis["lon"] = axis = ax1.new_floating_axis(60,70)
        ax1.grid(True)


#        tr = PolarAxes.PolarTransform()
#        pi = npy.pi
#        angle_ticks = [npy.linspace(0,2*pi+pi/24,24)]
#        grid_locator1 = FixedLocator([v for v, s in angle_ticks])
#        tick_formatter1 = DictFormatter(dict(angle_ticks))
#        grid_locator2 = MaxNLocator(2)
#        grid_helper = floating_axes.GridHelperCurveLinear(tr, extremes=(-pi,pi,70,90),
#          grid_locator1=grid_locator1, grid_locator2=grid_locator2,
#          tick_formatter1 = tick_formatter1,
#          tick_formatter2=None)
#        ax1 = floating_axes.FloatingSubplot(fig1,

        ax.append(fig1.add_axes(addaxes,projection='polar'))
      else:
        ax.append(fig1.add_axes(addaxes))
        pyp.locator_params(axis='x')#,nticks=4)
        pyp.locator_params(axis='y')#,nticks=4)

        if rm_labels:
            if xstart != -1:
              ax[-1].set_xticks(npy.linspace(xstart,xend,(xend+xstep-xstart)/xstep))
              ax[-1].set_xlim(xstart-xstep/2,xend+xstep/2)
            if ystart != -1:
              ax[-1].set_yticks(npy.linspace(ystart,yend,(yend+ystep-ystart)/ystep))
              ax[-1].set_ylim(ystart-ystep/2,yend+ystep/2)
            if y != 0:
              ax[-1].set_xticklabels('')
            #  ax[-1].set_xticks(npy.linspace(xstart,xend,(xend+xstep-xstart)/xstep))
            #  ax[-1].set_xlim(xstart-xstep/2,xend+xstep/2)
            else:
              if xstart != xend:
            #    ax[-1].set_xticks(npy.linspace(xstart,xend,(xend+xstep-xstart)/xstep))
            #    ax[-1].set_xlim(xstart-xstep/2,xend+xstep/2)
                ax[-1].set_xlabel(xlabel)
            if x != 0:
              ax[-1].set_yticklabels('')
            #  ax[-1].set_yticks(npy.linspace(ystart,yend,(yend+ystep-ystart)/ystep))
            #  ax[-1].set_ylim(ystart-ystep/2,yend+ystep/2)
            else:
              ax[-1].set_ylabel(ylabel)
              if ystart != yend:
                ax[-1].set_yticks(npy.linspace(ystart,yend,(yend+ystep-ystart)/ystep))
                ax[-1].set_ylim(ystart-ystep/2,yend+ystep/2)
        else:
          ax[-1].set_xlabel(xlabel)
          ax[-1].set_ylabel(ylabel)
      loop_cnt = loop_cnt*1 + 1

  return ax, fig1
```

```python
def get_export(this_sect):
    
    if do_Moore:
        this_sect_in = this_sect
        this_sign = Moore_signs[Moore.sections.index(this_sect_in),0]
    else:
        if this_sect == "CA_West_81":
            this_sect_in = "CA_West"
        else:
            this_sect_in = this_sect
        this_sign = Bf_signs[Bf.sections.index(this_sect_in),0]
    
    ## minus before them in order to get +ve in
    A = -get_sect_volexp(this_sect)*this_sign
    B = -get_sect_volexp(this_sect,vartype='totSI')*this_sign
    C,t_month_= get_monthly_export(A,tvalue)
    D,t_month = get_monthly_export(B,tvalue)
    E,t_year_= get_yearly_export(A,tvalue)
    F,t_year = get_yearly_export(B,tvalue)
    
    G = -get_sect_areaexp(this_sect)*this_sign
    H = -get_sect_areaexp(this_sect,vartype='totSI')*this_sign
    J,t_month = get_monthly_export(G,tvalue)
    K,t_month = get_monthly_export(H,tvalue)
    L,t_year = get_yearly_export(G,tvalue)
    M,t_year = get_yearly_export(H,tvalue)
    
    return A,B,C,D,E,F,G,H,J,K,L,M,t_month,t_year

def get_monthly_export(field,tv):
    
    mint = tv[0].year
    maxt = tv[-1].year
    field_monthly = npy.zeros([12*(maxt-mint+1),1])
    t_monthly = []
    
    mcnt = 0
    for yy in range(mint,maxt+1):
        for mm in range(1,13):
            ml = 31
            if mm in [4,6,9,11]:
                ml = 30
            elif mm == 2:
                if npy.mod(yy,1972) == 0:
                    ml = 29
                else:
                    ml = 28
            ind = npy.nonzero((tvalue>=dt.datetime(yy,mm,1,0,0,0)) & (tvalue<dt.datetime(yy,mm,ml,23,59,59)))
            field_monthly[mcnt,0] = npy.nansum(field[ind])/4 ## m/day to m/6h
            t_monthly = npy.append(t_monthly,dt.datetime(yy,mm,15,0,0,0))
            mcnt = mcnt + 1
            
    return field_monthly, t_monthly

def get_yearly_export(field,tv):
    
    mint = tv[0].year
    maxt = tv[-1].year
    field_yearly = npy.zeros([(maxt-mint+1),1])
    t_yearly = []
    
    mcnt = 0
    for yy in range(mint,maxt+1):

            ind = npy.nonzero((tvalue>=dt.datetime(yy,1,1,0,0,0)) & (tvalue<dt.datetime(yy,12,31,23,59,59)))
            field_yearly[mcnt,0] = npy.nansum(field[ind])/4 ## m/day to m/6h
            t_yearly = npy.append(t_yearly,dt.datetime(yy,1,1,0,0,0))
            mcnt = mcnt + 1
            
    return field_yearly, t_yearly

def plot_subplots(ax1,ax2,ax3,ax4,this_myi,this_sic,this_myi_month,this_sic_month,this_myi_year,this_sic_year,t_m,t_y,varname,thissect,units=1,
                  winter_m=[1,2,3],do_cum_month=0,do_FYI=1,output_vals=0):
    
    if do_cum_month > 0:
        do_myi_month = npy.zeros(npy.shape(this_myi_month[:,0]))
        do_fyi_month = npy.zeros(npy.shape(this_myi_month[:,0]))
        ncnt = 0
        for nn in range(0,e_year-s_year+1):
            do_myi_month[ncnt:ncnt+do_cum_month] = npy.cumsum(this_myi_month[ncnt:ncnt+do_cum_month,0]/units)
            do_fyi_month[ncnt:ncnt+do_cum_month] = npy.cumsum((this_sic_month[ncnt:ncnt+do_cum_month,0]-this_myi_month[ncnt:ncnt+do_cum_month,0])/units)

            ncnt = ncnt + 12
    else:
        do_myi_month = npy.squeeze(this_myi_month)/units;
        do_fyi_month = npy.squeeze(this_sic_month)/units-npy.squeeze(this_myi_month)/units
    
    month_neg_myi = do_myi_month*1; month_neg_myi[month_neg_myi>0] = 0
    month_pos_myi = do_myi_month*1; month_pos_myi[month_pos_myi<0] = 0
    if do_FYI:
        month_neg_fyi = do_fyi_month*1; month_neg_fyi[month_neg_fyi>0] = 0
        month_pos_fyi = do_fyi_month*1; month_pos_fyi[month_pos_fyi<0] = 0    
    else:
        month_neg_fyi = 0
        month_pos_fyi = 0
        
    if (plot_all_cumul_or_end == 'all') | (do_cum_month==0):
        ax1.bar(npy.arange(t_y[0].year+1./24,t_y[-1].year+1,1./12),month_neg_myi,color=colour_myi,width=2*wd/12)
        ax1.bar(npy.arange(t_y[0].year+1./24,t_y[-1].year+1,1./12),month_neg_fyi,bottom=month_neg_myi,color=colour_fyi,width=2*wd/12)
        ax1.bar(npy.arange(t_y[0].year+1./24,t_y[-1].year+1,1./12),month_pos_myi,color=colour_myi,width=2*wd/12)
        ax1.bar(npy.arange(t_y[0].year+1./24,t_y[-1].year+1,1./12),month_pos_fyi,bottom=month_pos_myi,color=colour_fyi,width=2*wd/12)
    else:
        nrange = npy.arange(do_cum_month-1,len(month_neg_myi),12)
        month_neg_myi[npy.isnan(month_neg_myi)] = 0
        month_pos_myi[npy.isnan(month_neg_myi)] = 0
        month_neg_fyi[npy.isnan(month_neg_myi)] = 0
        month_pos_fyi[npy.isnan(month_neg_myi)] = 0
        ax1.bar(npy.arange(t_y[0].year,t_y[-1].year+1),month_neg_myi[nrange],color=colour_myi,width=2*wd)
        #if npy.ndim(month_neg_fyi) > 1:
        ax1.bar(npy.arange(t_y[0].year,t_y[-1].year+1),month_neg_fyi[nrange],bottom=month_neg_myi[nrange],color=colour_fyi,width=2*wd)
        ax1.bar(npy.arange(t_y[0].year,t_y[-1].year+1),month_pos_myi[nrange],color=colour_myi,width=2*wd)
        #if npy.ndim(month_pos_fyi) > 1:
        ax1.bar(npy.arange(t_y[0].year,t_y[-1].year+1),month_pos_fyi[nrange],bottom=month_pos_myi[nrange],color=colour_fyi,width=2*wd)
 
    ax1.legend(['MYI','FYI'])
    
    if thissect == "Net":
        myi_av = npy.nanmean(do_myi_month[npy.arange(do_cum_month-1,len(do_myi_month),12)])
        fyi_av = npy.nanmean(do_fyi_month[npy.arange(do_cum_month-1,len(do_myi_month),12)])
        ax1.plot(npy.arange(t_y[0].year,t_y[-1].year+1,1./12),myi_av*npy.ones(npy.shape(month_neg_myi)),'--',color=colour_myi)
        ax1.plot(npy.arange(t_y[0].year,t_y[-1].year+1,1./12),fyi_av*npy.ones(npy.shape(month_neg_myi)),'--',color=colour_fyi)
    else:
        myi_av = npy.nanmean(do_myi_month[npy.arange(do_cum_month-1,len(do_myi_month),12)])
        si_av = npy.nanmean((do_myi_month+do_fyi_month)[npy.arange(do_cum_month-1,len(do_myi_month),12)])
        ax1.plot(npy.arange(t_y[0].year,t_y[-1].year+1,1./12),myi_av*npy.ones(npy.shape(month_neg_myi)),'--',color=colour_myi)
        ax1.plot(npy.arange(t_y[0].year,t_y[-1].year+1,1./12),si_av*npy.ones(npy.shape(month_neg_myi)),'--',color='r')
        
    neg_myi = npy.squeeze(this_myi_year)/units; neg_myi[neg_myi>0] = 0
    pos_myi = npy.squeeze(this_myi_year)/units; pos_myi[pos_myi<0] = 0
    if do_FYI:
        neg_fyi = npy.squeeze(this_sic_year)/units-npy.squeeze(this_myi_year)/units; neg_fyi[neg_fyi>0] = 0
        pos_fyi = npy.squeeze(this_sic_year)/units-npy.squeeze(this_myi_year)/units; pos_fyi[pos_fyi<0] = 0 
        this_fyi_year = this_sic_year-npy.squeeze(this_myi_year)
    else:
        neg_fyi = 0
        pos_fyi = 0
        this_fyi_year = 0
        
    if npy.ndim(this_myi_year) > 1:
        myiplot = this_myi_year[:,0]*1
        fyiplot = this_fyi_year[:,0]*1
    else:
        myiplot = this_myi_year*1
        fyiplot = this_fyi_year*1
    ax3.bar(npy.arange(t_y[0].year-0.5*wd,t_y[-1].year+1-0.5*wd),myiplot/units,color=colour_myi,width=wd)
    ax3.bar(npy.arange(t_y[0].year+0.5*wd,t_y[-1].year+1),fyiplot/units,color=colour_fyi,width=wd)
    #ax3.bar(npy.arange(t_y[0].year,t_y[-1].year+1),pos_myi,color=colour_myi)
    #ax3.bar(npy.arange(t_y[0].year,t_y[-1].year+1),pos_fyi,bottom=pos_myi,color=colour_fyi)
    ax3.plot([t_y[0].year-0.5,t_y[-1].year+1.5],[0,0],'k')
    ax3.plot([t_y[0].year,t_y[-1].year+1],[npy.nanmean(npy.squeeze(this_sic_year)/units-npy.squeeze(this_myi_year)/units),npy.nanmean(npy.squeeze(this_sic_year)/units-npy.squeeze(this_myi_year)/units)],'--',color=colour_fyi)
    ax3.plot([t_y[0].year,t_y[-1].year+1],[npy.nanmean(npy.squeeze(this_myi_year)/units),npy.nanmean(npy.squeeze(this_myi_year)/units)],'--',color=colour_myi)
    
    ## Now winter!
    winter_myi_vals = npy.zeros([e_year-s_year+1,1])
    winter_sic_vals = npy.zeros([e_year-s_year+1,1])
    #print('winter vals should be shape ')
    #print(npy.shape(winter_myi_vals))
    for nn in winter_m:
        print('doing month '+str(nn))
        if winter_m[0] <= winter_m[-1]:
            ## winter start is same year as winter end
            print('same year')
            print(npy.shape(this_myi_month))
            winter_myi_vals[:,0] = (winter_myi_vals[:,0] + npy.squeeze(this_myi_month[npy.arange(nn-1,len(this_myi_month),12)]))
            winter_sic_vals[:,0] = (winter_sic_vals[:,0] + npy.squeeze(this_sic_month[npy.arange(nn-1,len(this_myi_month),12)]))
        elif winter_m[0] > winter_m[-1]:
            ## winter start is year before winter end (e.g. Dec - Mar)
            print('different year')
            if nn < 9:
                winter_myi_vals[1::,0] = winter_myi_vals[1::,0] + npy.squeeze(this_myi_month[npy.arange(12+nn-1,len(this_myi_month),12),0])
                winter_sic_vals[1::,0] = winter_sic_vals[1::,0] + npy.squeeze(this_sic_month[npy.arange(12+nn-1,len(this_myi_month),12),0])
            else:
                winter_myi_vals[1::,0] = winter_myi_vals[1::,0] + npy.squeeze(this_myi_month[npy.arange(nn-1,len(this_myi_month)-12,12),0])
                winter_sic_vals[1::,0] = winter_sic_vals[1::,0] + npy.squeeze(this_sic_month[npy.arange(nn-1,len(this_myi_month)-12,12),0])
    #print('winter vals is shape ')
    #print(npy.shape(winter_myi_vals))
    if do_FYI:
        winter_fyi_vals = winter_sic_vals - winter_myi_vals
    else:
        winter_fyi_vals = 0*winter_sic_vals
        
    winter_neg_myi = npy.squeeze(winter_myi_vals)/units; winter_neg_myi[winter_neg_myi>0] = 0
    winter_pos_myi = npy.squeeze(winter_myi_vals)/units; winter_pos_myi[winter_pos_myi<0] = 0
    winter_neg_fyi = npy.squeeze(winter_fyi_vals)/units; winter_neg_fyi[winter_neg_fyi>0] = 0
    winter_pos_fyi = npy.squeeze(winter_fyi_vals)/units; winter_pos_fyi[winter_pos_fyi<0] = 0    
  
    ax4.bar(npy.arange(t_y[0].year-0.5*wd,t_y[-1].year+1-0.5*wd),winter_myi_vals[:,0]/units,color=colour_myi,width=wd)
    ax4.bar(npy.arange(t_y[0].year+0.5*wd,t_y[-1].year+1),winter_fyi_vals[:,0]/units,color=colour_fyi,width=wd)
    ax4.legend(['MYI','FYI'])
    #ax4.bar(npy.arange(t_y[0].year,t_y[-1].year+1),winter_pos_myi,color='turquoise')
    #ax4.bar(npy.arange(t_y[0].year,t_y[-1].year+1),winter_pos_fyi,bottom=winter_pos_myi,color='violet')
    ax4.plot([t_y[0].year-0.5,t_y[-1].year+1.5],[0,0],'k')
    ax4.plot([t_y[0].year,t_y[-1].year+1],[npy.nanmean(winter_fyi_vals[:,0]/units),npy.nanmean(winter_fyi_vals[:,0]/units)],'--',color=colour_fyi)
    ax4.plot([t_y[0].year,t_y[-1].year+1],[npy.nanmean(winter_myi_vals[:,0]/units),npy.nanmean(winter_myi_vals[:,0]/units)],'--',color=colour_myi)

    ax1.set_title(thissect+' '+varname)
    ax2.set_title(thissect+' '+varname)
    ax3.set_title(thissect+' '+varname)
    ax4.set_title(thissect+' '+varname)
    
    if "divided by" in varname:
        print('no limits applied')
    elif "volume" in varname:
        #ax1.set_ylim([-20,20])
        ax2.set_ylim([-100,100])
        if thissect == "Net":
            minval = -300; maxval = 300;
        elif npy.nanmean(winter_myi_vals) < 0:
            minval = -300; maxval = 30;
        else:
            minval = -30; maxval = 300;
        ax3.set_ylim([minval,maxval])
        if thissect == "Net":
            minval = -200; maxval = 200;
        elif npy.nanmean(winter_myi_vals) < 0:
            minval = -200; maxval = 30;
        else:
            minval = -30; maxval = 200;[npy.arange(do_cum_month,len(month_neg_myi),12)]
        ax4.set_ylim([minval,maxval])
        for ax in [ax1,ax2,ax3,ax4]:
            ax.set_ylabel('km$^3$')
        ax1.set_ylabel
    elif "area" in varname:
        #ax1.set_ylim([-10,10])
        ax2.set_ylim([-50,50])
        if thissect == "Net":
            minval = -225; maxval = 225;
        elif npy.nanmean(winter_myi_vals) < 0:
            minval = -225; maxval = 22.5;
        else:
            minval = -22.5; maxval = 225;
        ax3.set_ylim([minval,maxval])
        if thissect == "Net":
            minval = -200; maxval = 200;
        elif npy.nanmean(winter_myi_vals) < 0:
            minval = -200; maxval = 22.5;
        else:
            minval = -22.5; maxval = 200;
        ax4.set_ylim([minval,maxval])
        for ax in [ax1,ax2,ax3,ax4]:
            ax.set_ylabel('x 10$^3$ km$^2$')
    for ax in [ax1,ax2,ax3,ax4]:
        ax.grid()
    
    ax1.set_xticks(npy.arange(t_y[0].year,t_y[-1].year+1,1))
    
    ax3.set_xticks(npy.arange(t_y[0].year,t_y[-1].year+1,2))
    ax4.set_xticks(npy.arange(t_y[0].year,t_y[-1].year+1,2))
                                      
    npy.savez(outdir+"Beaufort_exports_yearly.npz",MYI_export=neg_myi,MYI_import=pos_myi,FYI_export=neg_fyi,FYI_import=pos_fyi,year_arr=t_y)
    npy.savez(outdir+"Beaufort_exports_"+winter_str+".npz",MYI_export=winter_neg_myi,MYI_import=winter_pos_myi,FYI_export=winter_neg_fyi,FYI_import=winter_pos_fyi,time_arr=t_y)
    npy.savez(outdir+"Beaufort_exports_monthly.npz",MYI_export=month_neg_myi,MYI_import=month_pos_myi,FYI_export=month_neg_fyi,FYI_import=month_pos_fyi,time_arr=t_m)
    
    if output_vals:
        return winter_myi_vals,winter_fyi_vals
    else:
        return
```
