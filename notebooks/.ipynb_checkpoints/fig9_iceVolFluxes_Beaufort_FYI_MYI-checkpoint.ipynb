## Ice Transport 

```python
ds2 = xr.open_dataset('../data_input/Exports_MooreBox_JFMAMJ_cumul.nc') 
ds2
# the dataset is the cumulative ice volume and area transports in the Beaufort Sea (moore22 box) for each month from January through June 2000-2018. 
# Monthly values are centered around the 15th. 
```

## Calculate cumulative volume fluxes from January 1 - June 1

```python
## Cumulative fluxes from January 1 to June 1st (cool season)
sivFlux_JFM = []
sivFlux_JFMAM = []

fyiFlux_JFMAM = []
myiFlux_JFMAM = []

for yr,da in ds2.groupby('time.year'):
    
    sivFlux_JFM.append(da['SIvolume_Net'].sel(time=da.time.dt.month.isin([1, 2, 3])).cumsum(axis=0).values[-1]) # by April 1 (end of winter)

    sivFlux_JFMAM.append(da['SIvolume_Net'].sel(time=da.time.dt.month.isin([1, 2, 3,4,5])).cumsum(axis=0).values[-1]) # by June 1 (end of cool season)
    
    myiFlux_JFMAM.append(da['MYvolume_Net'].sel(time=da.time.dt.month.isin([1, 2, 3,4,5])).cumsum(axis=0).values[-1]) # by June 1 (end of cool season)
    
```

## Plot Figure 9

```python
# create dataframe
years = np.unique(ds2.time.dt.year)
SIflux = pd.DataFrame({
    "SIvolFlux_JFM":np.array(sivFlux_JFM)*1E-2, # in 10^-2 km3
    "SIvolFlux_JFMAM":np.array(sivFlux_JFMAM)*1E-2,
    "MYvolFlux_JFMAM":np.array(myiFlux_JFMAM)*1E-2, 
    "FYvolFlux_JFMAM":(np.array(sivFlux_JFMAM) - np.array(myiFlux_JFMAM))*1E-2, 
    }, 
    index=years
)

ax = SIflux.SIvolFlux_JFMAM.plot(marker='o',ls='None', markersize=5, color='black',zorder=2,  use_index=False, figsize=(6,2.5))
SIflux.plot(kind='bar', y= ['FYvolFlux_JFMAM'], alpha=0.5, ax=ax)
SIflux.plot(kind='bar', y= 'MYvolFlux_JFMAM', color='tab:orange', alpha=0.5, ax=ax, rot=0)

plt.legend(['Net ice flux', 'FYI', 'MYI'], ncols=3)
plt.grid(ls='--', alpha=0.4)
plt.ylabel(r'Ice volume flux on June 1' "\n" r'($\times10^{-2}$ km$^3$)')

xticks = ax.xaxis.get_major_ticks()
for i,tick in enumerate(xticks):
    if i%2 != 0:
        tick.label1.set_visible(False)
        
plt.show()
fig = ax.get_figure()
fig.savefig('../figures/' + f'fig09_icevolFlux_JFMAM_moore22_2000-2018.png', dpi=300, bbox_inches='tight')
```
