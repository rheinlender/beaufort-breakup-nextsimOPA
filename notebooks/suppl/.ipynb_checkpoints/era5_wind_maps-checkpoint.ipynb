## Plot atmospheric synoptic conditions from ERA5 data

```python
%matplotlib inline
import sys, os
import xarray as xr
import datetime as dt
import pandas as pd
import numpy as np
import cmocean
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.gridspec as gridspec
from matplotlib import colors
import cartopy.crs as ccrs
import cartopy.feature as cfeature
sys.path.append('../../src')
from pynextsim.projection_info import ProjectionInfo
from brkup_utils import lead_detect as lead_det
from brkup_utils import process_data
from brkup_utils import mask_funcs as msk
from brkup_utils import grid_funcs as grd 
from brkup_utils.openerEra5 import OpenerEra5
from brkup_utils.boxnames import *
```

```python
# set environment
os.environ['INPUT_OBS_DATA_DIR'] = '/home/rheinlender/shared-simstore-ns9829k/data/'
os.environ['NEXTSIM_MESH_DIR'] = os.environ['INPUT_OBS_DATA_DIR'] + 'mesh/'

# set directories
#era5_dir = '/home/rheinlender/shared-simstore-ns9829k/data/ERA5/'    
INDIR = '/home/rheinlender/shared-simstore-ns9829k/NANUK/NANUK025-ILBOXE140-S/' 

# Load daily ERA5 data
varname=['u10', 'v10', 'msl']
year0 = 2000
year1 = 2018

era_file = []
for var in varname:
    # create the opener
    op = OpenerEra5(var)

    for year in range(year0, year1+1):

        # get the file name
        dto = dt.datetime(year,1,1)
        era_file.append(op.find(dto))

ds = xr.open_mfdataset(era_file, parallel=True, engine='netcdf4',chunks={'time':-1,'y': 300, 'x':250})

# get lon/lat
lat_e = ds.latitude.values
lon_e = ds.longitude.values
lat_e = lat_e[::-1] # make sure latitude is increasing    
    
# get winter months from daily data
months=[1,2,3]
ds_JFM = ds.sel(time = np.in1d(ds['time.month'], months), drop=True)
```

```python
# create BS mask 

lon=ds.longitude
lat=ds.latitude

msk_moore22 = 0*np.ones(ds.u10[0].shape)
msk_moore22[(lat<78)&(lat>69)&(lon>(360-160))&(lon<(360-120))] = 1
```

```python
# calculate 10-m wind speed
wspeed = np.hypot(ds_JFM.u10, ds_JFM.v10)

ds_JFM['windspeed'] = (['time', 'latitude', 'longitude'], wspeed.values)
```

```python
# test that it looks okay in native projection

# get lon/lat
lat_e = ds.latitude.values
lon_e = ds.longitude.values
#lat_e = lat_e[::-1] # make sure latitude is increasing    
    
# Set the projection to use for plotting
fig = plt.figure(1, figsize=[20,10])
ax = plt.subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()

# Pass ax as an argument when plotting. Here we assume data is in the same coordinate reference system than the projection chosen for plotting
# isel allows to select by indices instead of the time values
ds_JFM['windspeed'].sel(time='2001-01').mean('time').plot.pcolormesh(ax=ax, cmap='coolwarm', levels = np.arange(0,11,1), cbar_kwargs={'location':'bottom', 'shrink':0.4})

ax.set_extent([-160, -110, 67, 80], ccrs.PlateCarree())
ax.add_feature(cfeature.LAND, edgecolor='black', zorder=1, alpha=0)
ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
```

# Plot maps of ERA5 winds and SLP (Polar Stereographic)

```python
def plot_winds(ds, ax, cargs, qargs): 
    
    lat_e = ds.latitude.values
    lon_e = ds.longitude.values
    
    lon2D, lat2D = np.meshgrid(lon_e, lat_e)

    step = 2

    x = lon2D[::step, ::step]
    y = lat2D[::step, ::step]
    u = ds.u10[::step, ::step].values
    v = ds.v10[::step, ::step].values

    u_norm = u / np.sqrt(u ** 2.0 + v ** 2.0)
    v_norm = v / np.sqrt(u ** 2.0 + v ** 2.0)

    im=ax.contourf(lon_e, lat_e, ds.windspeed, transform=ccrs.PlateCarree(), **cargs)
    
    q = ax.quiver(x,y,u,v, transform=ccrs.PlateCarree(), **qargs)
    
    return im,q
    
def plot_slp(ds, ax, cargs): 
    
    lat_e = ds.latitude.values
    lon_e = ds.longitude.values
    
    lon2D, lat2D = np.meshgrid(lon_e, lat_e)

    msl = ds.msl*1E-2
    
    c = ax.contour(lon_e, lat_e, msl, transform=ccrs.PlateCarree(), **cargs)
   
    return c
    
month_arr = [1,2,3]
#year_arr = np.arange(2001, 2019, 1)
year_arr = [2008, 2010, 2013, 2016, 2018, 2005]

labels = ('January', 'February', 'March')
cargs = {'levels': np.arange(0,11,1),
         'cmap':'coolwarm' , 
         'extend':'max'}

qargs = {'regrid_shape':15,
         'scale':100}

slp_args = {'levels':20,
            'colors':'lightseagreen', 
            'alpha':0.7, }

# Set the projection to use for plotting
plot_crs = ccrs.NorthPolarStereo(central_longitude=-140, true_scale_latitude=60)

for yr in year_arr:
    
    fig, axs = plt.subplots(1,3,figsize=(12,8), subplot_kw={'projection': plot_crs}, layout='constrained')
    
    kw= {'edgecolor':"black", 'linewidth':0.5}
    
    for i, ax, label in zip(month_arr, axs.flat, labels):

        dto = f'{yr}-0{i}'
        print(dto)

        # Extract transect 
        ds_sel = ds_JFM.sel(time=dto).mean('time')
        
        # Set axes prop
        ax.set_extent([-160, -120, 68, 78], ccrs.PlateCarree())
        ax.add_feature(cfeature.LAND, edgecolor='black', zorder=1, alpha=0)
        ax.coastlines(resolution="50m",linewidth=1)
        
        # Add gridlines
        gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=False,
                          linewidth=0.5, color='black', linestyle='--', alpha=0.4)
        gl.top_labels = False
        gl.right_labels = False
        gl.bottom_labels = False
        gl.left_labels = False
        gl.xlines = False
        #gl.xlocator = mticker.FixedLocator([-65,-60,-50,-40,-30])
        #gl.ylocator = mticker.FixedLocator([30,40,50,60])
        #gl.xformatter = LONGITUDE_FORMATTER
        #gl.yformatter = LATITUDE_FORMATTER
        #gl.xlabel_style = {'size':10, 'color':'black'}
        #gl.ylabel_style = {'size':10, 'color':'black'}

        # Plot windspeed            
        im, q = plot_winds(ds_sel, ax, cargs, qargs)
        
        # Plot Sea level pressure contours
        cont = plot_slp(ds_sel, ax, slp_args)
        
        ax.text(.02, .98, label, ha='left', va='top', transform=ax.transAxes, fontsize=12, fontweight='bold', 
               bbox=dict(facecolor='white', edgecolor='black', pad=2.0, alpha=.7)
               )
        
        if i==2:
            ax.set_title(f'{yr}', rotation=0, fontsize=14, fontweight='bold')
            
    cb = plt.colorbar(im, orientation="vertical", aspect=20, shrink=0.28)
    cb.set_label('(m/s)', rotation=0, y=1.12, labelpad=-20)
            
    #fig.savefig(f'ERA5_wspeed_wdir_slp_maps_BS_{yr}.png', dpi=300, bbox_inches='tight')
```

## Calculate winds and SLP along 150W for the 2013 breakup event 

```python

def plot_wind_vectors(ax, ds, xint=1):
    
    time = ds['time']
    u_vec = ds['u10'].mean('latitude')
    v_vec = ds['v10'].mean('latitude')

    # calculate the magnitude
    scale=2
#    maxmag = max(wspeed_transect)/scale

    #fig, ax = plt.subplots(1)
    #ax.set_ylim(-maxmag, maxmag)
    q = ax.quiver(time[::xint], 0, u_vec[::xint],v_vec[::xint],width=0.003, scale=180)
    
    return q


month_arr = [1,2,3]
month_str = ['Jan', 'Feb', 'Mar']
#year_arr = np.arange(2013, 2019, 1)
year_arr = [2013]

import matplotlib.dates as mdates
days = mdates.DayLocator()  # every day
dtFmt = mdates.DateFormatter('%b %d') # define the formatting  

for yr in year_arr:
    fig, axs = plt.subplots(2,1,figsize=(5,4))
    
    # Extract transect 
    ds_slice = ds_JFM.sel(time=str(yr),
                         longitude=360-150, 
                         latitude=slice(75,70), 
                         drop=True)
    u = ds_slice.u10
    v = ds_slice.v10

    #-- Compute the wind speed (magnitude)
    wind_speed = np.sqrt(u**2 + v**2)
    slp_diff = (ds_slice.msl.sel(latitude=75) -  ds_slice.msl.sel(latitude=70)) *1E-2 
    
    slp_diffmean = slp_diff.mean('time')
    slp_diffstd = slp_diff.std('time')
    upper_lim = slp_diffmean.values + slp_diffstd.values
    
    #-- Compute the averages
    wind_speedavg = np.mean(wind_speed, axis=(-1))
    
    axs[0].plot(ds_slice.time.values, wind_speedavg)
    axs[0].axhline(y=0, ls='-', color='black', lw=1, alpha=0.4)
    q = plot_wind_vectors(axs[0], ds_slice, xint=2)
    
    axs[1].plot(ds_slice.time.values, slp_diff)
    axs[1].axhline(y=upper_lim, ls=':', color='black', lw=1)
    axs[1].axhline(y=0, ls='-', color='black', lw=1, alpha=0.4)

    # labels
    axs[0].set_ylim([-2,15])
    axs[1].set_ylim([-20,30])

    axs[0].set_title(str(yr),  y=1.0, loc='left', pad=-14, fontweight='bold')

    axs[0].set_ylabel('Wind speed (m s$^{-1}$)')
    axs[1].set_ylabel('$\Delta P$ (hPa)')
    
    for ax in axs.flat:
        ax.xaxis.set_major_formatter(dtFmt)
        #ax.xaxis.set_minor_locator(days+5)

    axs[0].set_xticklabels([])

    fig.subplots_adjust(bottom=0.01, top=0.98, left=0.02, right=0.86, wspace=0.04, hspace=0.08)

#    fig.savefig(figdir + f'ERA5_wspeed_slpdiff_150W_70-75N_{yr}.png', dpi=300, bbox_inches='tight')
```
