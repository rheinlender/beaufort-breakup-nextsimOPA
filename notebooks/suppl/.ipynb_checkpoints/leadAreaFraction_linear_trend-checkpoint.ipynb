# Calculating linear trends in Beaufort lead area fraction

```python
%matplotlib inline
import sys
import xarray as xr
import datetime as dt
import pandas as pd
import numpy as np
import cmocean
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.gridspec as gridspec

sys.path.append('../../src')
from pynextsim.projection_info import ProjectionInfo
from brkup_utils import lead_detect as lead_det
from brkup_utils import process_data
from brkup_utils import mask_funcs as msk
from brkup_utils import grid_funcs as grd 
from brkup_utils.openerEra5 import OpenerEra5
from brkup_utils.boxnames import *
```

```python
# read moorings 
indir = '/home/rheinlender/shared-simstore-ns9829k/NANUK/NANUK025-ILBOXE140-S/'    

months = ['01', '02', '03']
years = list(range(2000, 2019))
region='Beaufort'

# subset Beaufort Sea
ds = process_data.load_moorings(indir,months,years,region='Beaufort',period=None)
```

```python
# create masks
bbox = BOXNAMES[region]
mask_name='nsidc_Beaufort'
Mask = msk.Masking(ds, bbox=bbox)
mask_BS = Mask.get_nsidc_mask(mask_name, depth=None)['data']

msk_moore2022 = Mask.get_Moore2022_Beaufort_mask()

mskBS_d300 = Mask.get_nsidc_mask(mask_name, depth=300)['data']
```

```python
# calculate lead fraction
clim=0.05
leadfrac, leadmask = lead_det.map_leads(ds, 'breakup_paper', clim)
    
# daily lead area frac in Beaufort Sea
lfa = leadmask.where(msk_moore2022).mean(dim=('x', 'y'))
lfa_daily = lfa.groupby('time.date').mean()
```

```python
# group by months
year_month_idx = pd.MultiIndex.from_arrays([np.array(lfa['time.year']), 
                                            np.array(lfa['time.month'])])
lfa.coords['year_month'] = ('time', year_month_idx)

lfa_monthly = lfa.groupby('year_month').mean() # evarge over each year and month

lfa_JFMmean = lfa.groupby('time.year').mean('time')
lfa_JFMmean
```

```python

# create dataframe
df = pd.DataFrame({
    "lfa_jan": lfa_monthly.sel(time_level_1=1).values*100, # Jan
    "lfa_feb": lfa_monthly.sel(time_level_1=2).values*100, # Feb
    "lfa_mar": lfa_monthly.sel(time_level_1=3).values*100, # Mar
    "lfa_jfm":(lfa_JFMmean.values)*100, # JFM mean
    }, 
    index=years
)
```

```python
for key, data in df.items():

    _mean = data.mean(axis=0) #print(key, ': Average:', round(mean,2), 'STD:', round(std,2) )
    _std = data.std(axis=0) #print(key, ': Average:', round(mean,2), 'STD:', round(std,2) )
    
    print("{:s}: Average = {:.3f}%, STD = {:.3f}%".format(key, _mean, _std))
```

## Calculate Linear Trends

```python
## significance test

def plot_linearTrend(ax, data):
    
    from sklearn.linear_model import LinearRegression
    from sklearn.metrics import r2_score
    #%pip install pingouin
    import pingouin as pg

    key = data.name
    
    # input variables for linear regression
    X = np.array(data.index)
    Y = np.array(data)
    
    X = X.reshape(-1, 1)
    Y = Y.reshape(-1, 1).flatten()
    
    reg = LinearRegression()
    reg.fit(X,Y)
    params = np.append(reg.intercept_,reg.coef_)

    predictions = reg.predict(X)

    # get statistics from linear regression model (pengouin)
    reg_stats = pg.linear_regression(X , Y)

    #calculate R-squared of regression model
    r2 = reg_stats.r2[1]

    params = np.append(reg_stats.coef[0],reg_stats.coef[1])

    pval = reg_stats.pval[1]

    print("The linear model is: Y = {:.5} + {:.5}X".format(params[0], params[1]))
    print("r2:", r2)
    print("p-value:", pval)

    trend_lab = f'Trend: {np.round(params[1]*10, decimals=1)}% per decade'
    label = "r\u00b2 = {:.1f}% (p={:.3f})".format(r2*100, pval)
    
    ## Plotting
    ax.plot(X, Y, marker='o',linestyle='--',markersize=6)
    h1 = ax.plot(X, predictions, c='grey', linewidth=2, label=trend_lab)

    ax.grid( alpha=0.4, ls='--', zorder=1);
    #ax.set_title('Beaufort, JFM mean')
    ax.set_xticks(np.arange(2000, 2018+2,3))
    ax.legend(loc='upper left')

    # annotate
    
    ax.text(0.05, 0.1, label, transform=ax.transAxes, fontsize=11, va='top', color='grey') 

fig, axs = plt.subplots(2,2, figsize=(10,6))
labels = ('January lead fraction', 'February lead fraction', 'March lead fraction', 'JFM-mean lead fraction')
plt.subplots_adjust(hspace=0.3)
for (key,data), ax, lab in zip(df.items(), axs.ravel(), labels):
    ax.set_ylabel('Lead area fraction (%)')
    ax.set_title(lab)
    ax.set_ylim([5, 35])
    plot_linearTrend(ax, data)

plt.savefig('../../figures/' + 'figS01_leadFraction_Jan_Feb_Mar_JFM_moore22_linearTrend_2000-2018.png', dpi=300, bbox_inches='tight')
```
